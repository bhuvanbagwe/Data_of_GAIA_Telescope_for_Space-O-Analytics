```python
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
from matplotlib import image
from matplotlib import patches
from matplotlib import cbook
import numpy as np
```

```python
# CSV Location
df =pd.read_csv('C:/Users/BHUVAN/Desktop/Satellite Image Processing/Gaia/dataset.csv', header =None, nrows=140000, low_memory=False)
```

```python
# Giving Names 
distance = df[2]
radius = df[5]
luminosity = df[6]
mass=df[8]
g_latitude = df[11]
g_longitude = df[12]
temperature = df[15]
```

```python
with cbook.get_sample_data('C:/Users/BHUVAN/Desktop/Satellite Image Processing/Gaia/GaiaDR3StoryProperMotion_Gaia_EDR3_flux_hammer_small.jpg') as image_file:
    galaxy_image = plt.imread(image_file)

fig, ax = plt.subplots()
ax.plot(g_latitude,g_longitude,'.',color='white')

plt.title('Area of Interest : White Colored Patch')
plt.xlabel(' Galactic Longitude ')
plt.xlabel(' Galactic Latitude ')
im = ax.imshow(galaxy_image, extent=[0,360,-90,90]) # Setting the extent in range of Galactic Latitude and Longitude
plt.show()
```

```python
# Plot of Temperature vs Luminsoity (Mapped according to Mass)
# Resources :https://en.wikipedia.org/wiki/Hertzsprung%E2%80%93Russell_diagram 

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1)

logfmt = matplotlib.ticker.LogFormatterExponent(base=10, labelOnlyBase=True)
ax.set_yscale('log')

plt.gca().invert_xaxis() 
plt.scatter(temperature, luminosity,c=mass,cmap="plasma",marker='.')
plt.title('Temperature v/s Luminosity (Mapped according to Mass)')
plt.xlabel("Temperature (K)")
plt.ylabel("Luminosity")
plt.colorbar(label=" Range of Mass of Stars (solMass)", orientation="horizontal")
plt.savefig('C:/Users/BHUVAN/Desktop/Satellite Image Processing/Gaia/Temperature_vs_Luminosity_Mapped_according_to_Mass_.png', dpi=100)
plt.show()

```

```python
print("""
Above observations align closely with the Hertzsprung-Russell (H-R) Diagram of stars
Thereby, giving insights into the type of Stars, i.e SuperGiants, Giants, Main Sequnece stars and White dwarfs


About the Hertzsprung-Russell (H-R) Diagram : 
HertzsprungRussell diagram, abbreviated as HR diagram, HR diagram or HRD, 
is a scatter plot of stars showing the relationship between the stars absolute magnitudes or luminosities 
versus their stellar classifications or effective temperatures. """ )

hr = image.imread('C:/Users/BHUVAN/Desktop/Satellite Image Processing/Gaia/hr_plot.jpg')
hr_plot = plt.imshow(hr)
plt.title(" Hertzsprung-Russell (H-R) Diagram ")
plt.show()

print("""
The Hertzsprung-Russell (H-R) Diagram can be divided into certain prominent parts as follows 

1. Main Sequence (most stars)
About 85% of stars nearby Sun in our galaxy lie along a diagonal the band in the H-R Diagram, 
This band of stars is called the Main Sequence
Properties of Main Sequence : 
L=10-2 to 106 Lsun
T=3000 to >50,0000 K
R=0.1 to 10 Rsun

2. Giant & Supergiant Branches
There branches of stars in the H-R diagram are brighter than Main Sequence stars but have the same effective temperatures.

>> Luminosity-Radius-Temperature relation of these stars tells us that, stars in these bands are larger in radius than Main Sequence stars.
This can be clearly seen in the Temperature v/s Luminosity (Mapped according to Radius of Star) plot shown below 

There are two groups of giant stars:
Giants 
(cool stars with a wide range of luminosities)
R = 10 to 100 Rsun
L = 103 to 105 Lsun

Supergiants 
(largest stars, arranged along the top of the H-R diagram
with a wide range of effective temperatures but relatively narrow range of luminosities)
R > 103 Rsun
L = 105 to 106 Lsun

3. White Dwarfs
White dwarfs are found relatively less in numbers and occupy the lower left-hand corner of the H-R Diagram

Temperature v/s Luminosity (Mapped according to Radius of Star) plot below tells us that these stars 
are smaller in radius than Main Sequence stars.

""")
```
